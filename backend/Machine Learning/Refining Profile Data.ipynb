```python
import pandas as pd
import _pickle as pickle
import numpy as np
from scipy.stats import halfnorm
```

```python
with open("profiles.pkl",'rb') as fp:
    df = pickle.load(fp)
```

```python
# Removing the numerical data
df = df[['Bios']]
```

```python
# Probability dictionary
p = {}

# Languages
Languages = [
  'Afrikaans',
  'Arabic',
  'Chewa',
  'Hausa',
  'Igbo',
  'Oromo',
  'Somali',
  'Swahili',
  'Yoruba',
  'Zulu',
]

p['Languages'] = [0.28,
               0.21,
               0.16,
               0.14,
               0.09,
               0.06,
               0.04,
               0.01, 
               0.01]

# Music
Music = ['Afrobeat',
  'Benga',
  'Chimurenga',
  'Coupé-Decalé',
  'Fuji',
  'Isicathamiya',
  'Kizomba',
  'Makossa',
  'Mbalax',
  'Tcha tcho',]

p['Music'] = [0.30,
           0.23,
           0.12,
           0.12,
           0.09,
           0.08,
           0.03,
           0.02,
           0.01]

# Cuisine
Cuisine = ['Beinin Cuisine',
  'Burkinabé Cuisine',
  'Gambian Cusine',
  'Ghanaian Cuisine',
  'Ivorian Cuisine',
  'Liberian Cusine',
  'Malian Cuisine',
  'Nigerian Cuisine',
  'Senegalese Cuisine',
  'Togolese Cuisine',]

p['Cuisine'] = [0.16,
                 0.16,
                 0.01,
                 0.19,
                 0.11,
                 0.05,
                 0.10,
                 0.09,
                 0.07,
                 0.06]

# Arts
Arts = [ 'Dance',
  'Instruments',
  'Jewelry',
  'Painting',
  'Poetry',
  'Pottery',
  'Sculpting',
  'Singing',
  'Storytelling',
  'Textiles',]

p['Arts'] = [0.30,
              0.23,
              0.20,
              0.10,
              0.06,
              0.04,
              0.03,
              0.02,
              0.02]

# History
History = ['Beinin History',
  'Burkinabé History',
  'Gambian Cusine',
  'Ghanaian History',
  'Ivorian History',
  'Liberian Cusine',
  'Malian History',
  'Nigerian History',
  'Senegalese History',
  'Togolese History',]

p['History'] = [0.34,
               0.30,
               0.16, 
               0.13,
               0.04,
               0.03]

# Lists of Names and the list of the lists
categories = [Languages, Music, Cuisine, Arts, History]

names = ['Languages','Music', 'Cuisine', 'Arts', 'History']

combined = dict(zip(names, categories))
```

```python
# Looping through and assigning random values
for name, cats in combined.items():
    
        # Picking 3 from the list 
        try:
            df[name] = list(np.random.choice(cats, size=(df.shape[0],1,3), p=p[name]))
        except:
            df[name] = list(np.random.choice(cats, size=(df.shape[0],1,3)))
        
        df[name] = df[name].apply(lambda x: list(set(x[0].tolist())))
```
